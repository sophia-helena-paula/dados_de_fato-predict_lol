```python
import pandas as pd
import os
pd.set_option('display.max_column', None)
```

```python
champions_path = os.path.join('..', 'data', 'champs.csv')
matches_path = os.path.join('..', 'data', 'matches.csv')
participants_path = os.path.join('..', 'data', 'participants.csv')
stats1_path = os.path.join('..', 'data', 'stats1.csv')
stats2_path = os.path.join('..', 'data', 'stats2.csv')
teambans_path = os.path.join('..', 'data', 'teambans.csv')
teamstats_path = os.path.join('..', 'data', 'teamstats.csv')
```

```python
champions = pd.read_csv(champions_path)
matches = pd.read_csv(matches_path)
participants = pd.read_csv(participants_path)
stats1 = pd.read_csv(stats1_path)
stats2 = pd.read_csv(stats2_path)
teambans = pd.read_csv(teambans_path)
teamstats = pd.read_csv(teamstats_path)
```

```python
#name
#id
champions_output = champions.copy()
champions
```

```python
matches_output = matches[['id', 'gameid']].copy()
matches_output
```

```python
participants_output = participants[['id', 'matchid', 'player', 'championid', 'role', 'position']].copy()
participants_output
```

```python
stats = pd.concat([stats1, stats2], axis=0, ignore_index=True)
stats_output = stats[['id', 'win', 'kills']].copy()
stats_output
```

```python
teambans_output = teambans.copy()
teambans_output
```

```python
champions_output.to_csv(os.path.join('..', 'output_etl', 'champs.csv'))
matches_output.to_csv(os.path.join('..', 'output_etl', 'matches.csv'))
participants_output.to_csv(os.path.join('..', 'output_etl', 'participants.csv'))
stats_output.to_csv(os.path.join('..', 'output_etl', 'stats.csv'))
teambans_output.to_csv(os.path.join('..', 'output_etl', 'teambans.csv'))
```
